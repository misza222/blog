Based on [Pieter Abbeel lectures](https://www.youtube.com/playlist?list=PLwRJQ4m4UJjNymuBM9RdmB3Z9N5-0IlY0)

General rule: Sampling as a way to esitmate true (definitive) values in the infinite world of possibilities. So we don't get an erratic behaviour we introduce  (learning rate) to gradually update our estimates.



Greedy and -Greedy approaches to environment exploration

Q-learning converges to optimal policy. This even though we don't take optimal actions during training, but this exactly allows us to explore environment better than with Greedy approach. This is given you need to give algorithm enough time based on a state and action space of the environment. There are also technicalities that need to be considered - change  during training, so first we explore more, and as we approach convergence we focus more on tried and working paths.
